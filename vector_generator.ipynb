```python
import cv2         
import numpy as np
from matplotlib import pyplot as plt
```

```python
def turnTo01(fn):
    img = cv2.cvtColor(cv2.imread(fn), cv2.COLOR_BGR2GRAY)
    img[img == 0] = 1
    img[img != 1] = 0
    img = img.T
    print("std::vector<std::vector<int> > map = {} ".format('{'), end="\n")
    for i in range(0, len(img)):
        print("{", end=" ")
        for j in range(0, len(img[0])):
            if (j == len(img[0])-1):
                print(img[i][j], end="")
            else:
                print(img[i][j], end=",")
        if (i == len(img)-1):
            print("} };")
        else:
            print("},")
```

```python
turnTo01('./resources/sprites/collision_mask.png')
```
